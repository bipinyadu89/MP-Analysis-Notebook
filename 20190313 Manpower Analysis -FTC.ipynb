```python
import xlrd
import pandas as pd
from openpyxl import load_workbook
import datetime
```

```python
all=pd.DataFrame()
wb=load_workbook('manpower.xlsx')
for ws in wb.worksheets[180:210]:
    title=ws.title
    df=pd.read_excel('manpower.xlsx',sheet_name=ws.title,skiprows=2, usecols='A:J',na_values=0)
    df.columns=['#','Machine','SKU','STDMP','G','A','B','C','TOTAL','T']
    x=datetime.datetime.strptime(title,'%d-%m-%y')
    
    TL=[[title,df['TOTAL'].iloc[6],df['TOTAL'].iloc[32],df['TOTAL'].iloc[16],df['TOTAL'].iloc[14],df['TOTAL'].iloc[21]]]
    
    all=all.append(TL,ignore_index=True)
```
