```python
import xlrd
import pandas as pd
from openpyxl import load_workbook
wb = load_workbook('manpower.xlsx')
all=pd.DataFrame()
for ws in wb.worksheets[210:]:
    Title=ws.title
    df=pd.read_excel('manpower.xlsx',sheet_name=ws.title,skiprows=2,usecols='A:J',na_values=0)
    df.columns=['#','Machine','SKU','STDMP','G','A','B','C','TOTAL','REMARKS']
    TL=[[Title, df['TOTAL'].iloc[-1]]]
    all=all.append(TL,ignore_index= True)
all.columns=['Date','Manpower']
all
```
